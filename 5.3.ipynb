```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# Define the function given in the image
def func(x):
    return x**5 - 8*x**4 + 44*x**3 - 91*x**2 + 85*x - 26

# Create an array of x values from 0 to 1.5
x = np.linspace(0, 1.5, 400)

# Calculate the y values based on the function
y = func(x)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = x^5 - 8x^4 + 44x^3 - 91x^2 + 85x - 26')
plt.title('Graph of f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.legend()

# Find the root of the function using fsolve starting from a guess x=0.5
root = fsolve(func, 0.5)

# Plot the root
plt.scatter(root, [0], color='red')
plt.annotate(f'Root: ({root[0]:.2f}, 0)', xy=(root[0], 0), xytext=(10, -30),
             textcoords='offset points', arrowprops=dict(arrowstyle="->", color='red'))

# Show the plot
plt.show()

# Output the root as well
root
```
