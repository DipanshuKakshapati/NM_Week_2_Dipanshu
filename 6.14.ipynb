```python
import numpy as np
import matplotlib.pyplot as plt
```

### Newton Raphson Method

```python
# Redefine the function and derivative for plotting and applying methods
def f(x):
    return x**5 - 16.05*x**4 + 88.75*x**3 - 192.0375*x**2 + 116.35*x + 31.6875

def df(x):
    return 5*x**4 - 64.2*x**3 + 266.25*x**2 - 384.075*x + 116.35

# Redefine the methods for calculations
def newton_raphson_method(f, df, initial_guess, tolerance=0.01/100, max_iter=100):
    x_old = initial_guess
    history = [x_old]

    for _ in range(max_iter):
        x_new = x_old - f(x_old) / df(x_old)
        history.append(x_new)

        if abs(x_new - x_old) < tolerance:
            break
        
        x_old = x_new

    return history

# Apply the Newton-Raphson method with the initial guess
initial_guess = 0.5825
nr_history = newton_raphson_method(f, df, initial_guess)
nr_history_values = nr_history

# Prepare the plot for the Newton-Raphson Method
x_values = np.linspace(-2, 7, 400)
y_values = f(x_values)

plt.figure(figsize=(12, 6))
plt.plot(x_values, y_values, label='f(x)', color='blue')
plt.scatter(nr_history, [f(x) for x in nr_history], color='red', zorder=5)
plt.plot(nr_history, [f(x) for x in nr_history], label='Newton-Raphson Iterations', color='red', linestyle='--')
plt.title('Newton-Raphson Method')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

print('Iterations:' , nr_history_values)

# Output the number of iterations and t he final root approximation for reference
len(nr_history), nr_history[-1]
```

### Modified Secant Method

```python
# Redefine the function and derivative for plotting and applying methods
def f(x):
    return x**5 - 16.05*x**4 + 88.75*x**3 - 192.0375*x**2 + 116.35*x + 31.6875

def df(x):
    return 5*x**4 - 64.2*x**3 + 266.25*x**2 - 384.075*x + 116.35

def modified_secant_method(f, initial_guess, delta=0.05, tolerance=0.01/100, max_iter=100):
    x_old = initial_guess
    history = [x_old]

    for _ in range(max_iter):
        x_new = x_old - delta * x_old * f(x_old) / (f(x_old + delta * x_old) - f(x_old))
        history.append(x_new)

        if abs(x_new - x_old) < tolerance:
            break

        x_old = x_new

    return history

# Apply both methods with the initial guess
initial_guess = 0.5825
ms_history = modified_secant_method(f, initial_guess)
ms_history_values = ms_history

# Prepare plots for each method
x_values = np.linspace(-2, 7, 400)
y_values = f(x_values)

# Modified Secant Method Plot
plt.figure(figsize=(12, 6))
plt.plot(x_values, y_values, label='f(x)', color='blue')
plt.scatter(ms_history, [f(x) for x in ms_history], color='green', zorder=5)
plt.plot(ms_history, [f(x) for x in ms_history], label='Modified Secant Iterations', color='green', linestyle='--')
plt.title('Modified Secant Method')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

print('Iterations:' , ms_history_values)

# Output the number of iterations and t he final root approximation for reference
len(ms_history), ms_history[-1]
```
